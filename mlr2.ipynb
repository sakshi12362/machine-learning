```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as pt
```

```python

 x=np.array([140,150,160,170]).reshape(-1,1)
 y=np.array([60,70,80,90])
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
m=model.coef_[0] # slope
b=model.intercept_ # intercept
print(m)
print(b)
```

```python
x_new=np.array([20]).reshape(-1,1)
y_predict=model.predict(x_new)
print(f"{y_predict[0]:.2f}")
```

```python
pt.scatter(x,y,color='green',label="data point")
pt.plot(x,model.predict(x),color="blue",label="Regression Linear")
pt.xlabel("weight in lbs")
pt.ylabel("height in inches")
pt.title("height and weight distribution")
pt.legend()
pt.show()
```
