```python
import pandas as pd
```

```python
# https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data
# https://archive.ics.uci.edu/static/public/10/data.csv

# column
columns = [
    'symboling','normalized-losses','make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels','engine-location','wheel-base','length',
    'width','heigth','curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size','fuel-system','bore','stroke','compresion-ratio','horsepower','perk-rpm',
    'city-mpg','price'
]

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
data = pd.read_csv(url,names=columns, na_values='?', sep=',',skipinitialspace=True, engine='python')

# data.to_csv('auto_mobiles.csv')  # it is used to save the file in csv format
```

```python
data_unique =  ['make','fuel-type','aspiration','num-of-doors','bore','horsepower','body-style','drive-wheels','engine-location','num-of-cylinders',
                'fuel-system']
for col in data_unique:
    print(f'{col} : {data[col].unique()}')
```

```python
# fill the missing values
# 'aspiration','fuel-system','bore','horsepower','compresion-ratio '
```

```python
# handling missing data

mean_fuel_system = data['fuel-system'].mean()
data['fuel-system'].fillna(mean_fuel_system, inplace=True)
print(data['fuel-system'].head())
```

```python
mean_horsepower = data['horsepower'].mean()
data['horsepower'].fillna(mean_horsepower, inplace=True)
print(data['horsepower'].head())
```

```python
mean_price = data['price'].mean()
data['price'].fillna(mean_horsepower, inplace=True)
print(data['price'].head())
```

```python
mean_more = data['bore'].mean()
data['bore'].fillna(mean_more, inplace=True)
print(data['bore'].head())
```

```python
mean_compresion_ratio = data['compresion-ratio'].mean()
data['compresion-ratio'].fillna(mean_compresion_ratio, inplace=True)
print(data['compresion-ratio'].head())
```

```python
mode_aspirations = data['aspiration'].mode()[0]
data['aspiration'].fillna(mode_aspirations, inplace=True)
print(data['aspiration'].head())
```

```python
mode_symboling = data['symboling'].mean()
data['symboling'].fillna(mode_symboling, inplace=True)
print(data['symboling'].head())
```

```python
q1=data['horsepower'].quantile(0.25)
q3=data['horsepower'].quantile(0.75)
iqr=q3-q1
upper=q3+1.5*iqr
lower=q1-1.5*iqr
outliners=data[(data['horsepower']<lower)|(data['horsepower']> upper)]
print(outliners[['make','horsepower']])
print(lower)
print(upper)
print(iqr)
```
