```python
hours=np.array([[2],[4],[6]]).reshape(-1,1)
answer=np.array([0,1,1])



model=LogisticRegression()
model.fit(hours,answer)

print("starting point", model.intercept_[0])
print("coefficient", model.coef_[0][0])

guess = model.predict(hours)
print("Predicted values:", guess)
```
