```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as pt
```

```python
# data
x=np.array([8,10,12]).reshape(-1,1) # pizza diameter(independet variable)
y=np.array([10,11,12])# pizza price(dependent variable)
```

```python
# create and train the model
model=LinearRegression()
model.fit(x,y)
```

```python
# get the slope(m)
#get the intercept (b)
m=model.coef_[0] # slope
b=model.intercept_ # intercept
print(m)
print(b)
```

```python
# predict price for 20 inch =pricr?
x_new=np.array([20]).reshape(-1,1)
y_predict=model.predict(x_new)
print(f"{y_predict[0]:.2f}")
```

```python
pt.scatter(x,y,color='green',label="data point")
pt.plot(x,model.predict(x),color="blue",label="Regression Linear")
pt.xlabel("diameter")
pt.ylabel("price")
pt.title("pizza price distribution")
pt.legend()
pt.show()
```
